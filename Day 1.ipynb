## 数组

###  要点
数组是存储在连续内存空间的相同数据类型的集合



⚠️ 数组下标从0开始

⚠️ 内存空间连续

⚠️ 元素只能覆盖，不能删除

### 二分查找
双指针

```python
# 升序 数组nums 目标值target

class Solution:
    def search(self, nums, target):
        left = 0
        right = len(nums) - 1
        

        while left <= right:
            mid = (left + right) // 2
            if nums[mid] > target:
                right = mid - 1
            elif nums[mid] < target:
                left = mid + 1
            else:
                return mid
        return -1

nums = [-1,0,3,5,9,12]
target = 9
a = Solution()
print(a.search(nums, target))
```

### 元素移除



```python
# 双指针解决

class Solution:
    def remove(self, nums, target):
        slow = 0
        fast = 0

        while fast < len(nums):
            if nums[fast] != target:
                nums[slow] = nums[fast]
                slow += 1
            fast += 1
            
        return slow

nums = [3,2,2,3]
target = 3

a = Solution()
print(a.remove(nums, target))
```

### 有序数组的平方

```python
# 双指针

class Solution:
    def new_arr(self, nums):
        left = 0
        right = len(nums) - 1 
        new = [float('inf')] * len(nums)
        n = right

        while left <= right:
            if nums[left] ** 2 > nums[right] ** 2:
                new[n] = nums[left] ** 2
                left += 1
            else:
                new[n] = nums[right] ** 2
                right -= 1
            n -= 1
        return new

nums = [-7,-3,2,3,11]
a = Solution()
print(a.new_arr(nums))
```
