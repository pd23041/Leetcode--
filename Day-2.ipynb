### 长度最小的子数组



解决方法：1. 暴力解法 2. 滑动窗口

暴力解法思路： 两个for循环 时间复杂度n**2

滑动窗口思路： 双指针调节子序列起始位置 根据条件判断 slow和fast指针的移动 

```python
nums = [2,3,1,2,4,3]
target = 7

class Solution:
    def find_min(self, nums, target):
        slow = 0
        fast = 0
        results = []
        count = 0

        while fast < len(nums):
            count += nums[fast]
            while count >= target:
                count -= nums[slow]
                results.append(fast-slow+1)
                slow += 1

            fast += 1
        return min(results) if results else 0

a = Solution()
print(a.find_min(nums, target))
            
```

### 螺旋矩阵



解题思路： 主要涉及到去见定义问题 坚持循环不变原则 [左闭右开)

不同n对应不同层数的loop 



```python


class Solution:
    def get_matrix(self, n):
        x = 0
        y = 0
        matrix = [[float('inf')] * n for _ in range(n)] # 矩阵定义
        mid = n // 2
        loops = n // 2  # 循环圈数
        start = 1

        for offset in range(1, loops + 1):
            for j in range(y, n-offset):
                matrix[x][j] = start
                start += 1
            for j in range(x, n-offset):
                matrix[j][n-offset] = start
                start += 1
            for j in range(n-offset, y, -1):
                matrix[n-offset][j] = start
                start += 1
            for j in range(n-offset, x, -1):
                matrix[j][y] = start
                start += 1
        
        if n % 2 != 0:
            matrix[mid][mid] = start 
        return matrix

n = 3
a = Solution()
a.get_matrix(n)
```

## 区间和



主要是熟悉ACM 模式的输入 acm输入的是str格式的字符串 因此需要用int进行数据类型转换

```python
n = input()

sum_i = [float('inf')] * n

total = 0
for i in range(n):
    s = int(input())
    total += s
    sum_i = total

while True:
    try:
        a, b = map(int, input(),split())
        if a == 0:
            sum_ab = sum_i[b]
        
        else:
            sum_ab = sum_i[b] - sum_i[a-1]
        print(sum_ab)
        
        
    except:
        break
```
