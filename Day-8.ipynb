## 字符串

### 反转字符串



思路：数组只能覆盖，因此只能借助额外的temp进行覆盖，同时注意输入的收据类型，需要进行转换。

```python
class Solution(object):
    def reverseString(self, s):
        """
        :type s: List[str]
        :rtype: None Do not return anything, modify s in-place instead.
        """
        left = 0
        right = len(s) - 1
        while left < right:
            temp = s[right]
            s[right] = s[left]
            s[left] = temp
            left += 1
            right -= 1
        return s
        
```

### 反转字符串II



思路：对输入的str进行list类型转换，进行一个for循环，2k步数，同时对s[i:i+k]内的字符进行转换，最后输出''.join(s)输出成为字符串。

```python
class Solution(object):
    def reverseStr(self, s, k):
        """
        :type s: str
        :type k: int
        :rtype: str
        """
        
        s = list(s)

        for i in range(0, len(s), 2*k):
            s[i:i+k] = reversed(s[i:i+k])
        return ''.join(s)

    
        
   
```

### 替换数字



遍历数字，利用isdigit判断，然后替换成number，返回''.join()

```python
class Solution:
    def transfer(self, s):
        res = list(s)
        
        for i in range(len(res)):
            if res[i].isdigit():
                res[i] = 'number'
        return ''.join(res)
        
sol = Solution()
s = input()
result = sol.transfer(s)
print(result)
```
