## 字符串

### 反转字符串中的单词



思路：首先需要利用s.split对字符串s进行词分割，然后利用reverse的思路对词进行换位置，最后‘’.join（）



```python
class Solution(object):
    def reverseWords(self, s):
        """
        :type s: str
        :rtype: str
        """
        word = s.split()
        result = reversed(word)
        return ' '.join(result)
        
```

### 右旋转字符串



思路：可以将字符串转换成list，利用切片讲两个字符串进行相加。

注意事项：ACM模式下，int（input（））是获取第一行数字，input（）是获取list，map（int， int（）.split（））是获取数组里面的逐个元素。


```python
def reverse(s, k):
    a = len(s)-k
    s1 = s[:a]
    s2 = s[a:]
    return s2 + s1


k = int(input())
s = input()
result = reverse(s, k)
print(result)
```
