```python
class Solution(object):
    def reverseWords(self, s):
        """
        :type s: str
        :rtype: str
        """
        word = s.split()
        result = reversed(word)
        return ' '.join(result)
        
```

```python
def reverse(s, k):
    a = len(s)-k
    s1 = s[:a]
    s2 = s[a:]
    return s2 + s1


k = int(input())
s = input()
result = reverse(s, k)
print(result)
```
